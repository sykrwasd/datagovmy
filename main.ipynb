```python
import pandas as pd
```

```python
df = pd.read_csv("shapes.csv")
```

```python
df.head()
```

```python
df = df.sort_values(["shape_id", "seq"])
```

```python
grouped = (
    df.groupby("shape_id")
      .apply(lambda g: list(zip(g["lat"], g["lon"])))
      .to_dict()
)

# zip(g["lat"], g["lon"])


# zip(
#   [3.10378, 3.10377, 3.10377],
#   [101.64532, 101.64534, 101.64535]
# )

# (3.10378, 101.64532)
# (3.10377, 101.64534)
# (3.10377, 101.64535)
```

```python
import polyline

for items, i in grouped.items():
    print(items, polyline.encode(i))
```
